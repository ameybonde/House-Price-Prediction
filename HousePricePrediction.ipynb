```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
categorical_features=[i for i in x.columns if x[i].dtype=='O']
len(categorical_features)
```

```python
numerical_features=[i for i in x.columns if x[i].dtype!='O']
len(numerical_features)
```

```python
numerical_features
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor=ColumnTransformer(
    [
        ("OneHotEncoder",categorical_transformer,categorical_features),
        ("StandardScalar",numerical_transformer,numerical_features)
    ]
)
```

```python
x_train=preprocessor.fit_transform(x_train)
x_test=preprocessor.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

models = {
    "Linear": LinearRegression(),
    "RF": RandomForestRegressor(),
    "KNN": KNeighborsRegressor()
}
```

```python
for i in range(len(list(models))):
  model=list(models.values())[i]
  model.fit(x_train,y_train)

  y_train_pred=model.predict(x_train)
  y_test_pred=model.predict(x_test)

  mse_train = mean_squared_error(y_train, y_train_pred)
  mse_test = mean_squared_error(y_test, y_test_pred)


  print(model)
  print("Training Accuracy")
  print('mae:',mean_absolute_error(y_train, y_train_pred))
  print('mse:' , mse_train)
  print('rmse:' , np.sqrt(mse_train))
  print('r2:' , r2_score(y_train, y_train_pred))
  print()

  print(model)
  print("Testing Accuracy")
  print('mae:',mean_absolute_error(y_test, y_test_pred))
  print('mse:' , mse_test)
  print('rmse:' , np.sqrt(mse_test))
  print('r2:' , r2_score(y_test, y_test_pred))
  print()
```
